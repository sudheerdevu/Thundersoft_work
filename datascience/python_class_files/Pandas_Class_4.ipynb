# Cross Tabulation

```python
# - It is a method to analyse the multiple cataegorical columns.
# - It is used to draw the frequency distribution for the categorical column.
```

```python
data=['A','B','C','A','D','B','B','E']
fre_quen={'A':2,'B':3,'C':1,'D':1,'E':1}
```

```python
# pd.crosstab(cate,cate,aggfunc=,values=)
```

```python
data={'Product': ['A', 'A', 'B', 'B', 'C', 'C', 'A', 'B'],
'Region': ['North', 'South', 'North', 'South', 'North', 'South', 'South', 'North']}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
pd.crosstab(df.Product,df.Region)
```

```python
pd.crosstab(df.Product,df.Region,margins=True)
```

```python
sales=[100, 150, 200, 50, 300, 250, 200, 100]
```

```python
df['Sales']=sales
```

```python
df
```

```python
pd.crosstab(df.Product,df.Region,aggfunc='sum',values=df.Sales)
```

```python
pd.crosstab(df.Product,df.Region,aggfunc='sum',values=df.Sales,normalize='all')
```

```python
pd.crosstab(df.Product,df.Region,aggfunc='sum',values=df.Sales,normalize='index')
```

```python
pd.crosstab(df.Product,df.Region,aggfunc='sum',values=df.Sales,normalize='columns')
```

```python
pd.crosstab(df.Product,df.Region,aggfunc='sum',values=df.Sales,margins=True)
```
