```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.arange(1,6)
y=x**2
```

# Annotation

```python
plt.plot(x,y,
         color='green',
         marker='o',
         markersize=12,
         markerfacecolor='yellow',
        markeredgecolor='red',
        markeredgewidth=2,
        label='Sales Trend',
        lw=2.8,
        ls='--')

plt.title('Plot-1',fontdict={'color':'red','size':20,'font':'monospace'})
plt.xlabel('Array-X',fontdict={'color':'red','size':20,'font':'monospace'})
plt.ylabel('Array-Y',fontdict={'color':'red','size':20,'font':'monospace'})

plt.legend(fontsize=12,labelcolor='green')
plt.show()
```

```python
# plt.annotate(text,xy=(float,float),xytext=(float,float))
```

```python
# text- The text we want to display
# xy- Coordinates of something that we want to annotate
# xytext- Coordinates where we want to place our text
```

```python
plt.plot(x,y,marker='o',markersize=10)

for i,j in zip(x,y):
    plt.annotate(str(i)+','+str(j),
                 color='blue'
                 ,xy=(i,j),
                 xytext=(i-0.75,j+0.04),
                bbox=dict(facecolor='r',alpha=0.5,boxstyle='circle'),
                arrowprops=dict(arrowstyle='->',color='g'))

plt.show()
```

```python
for i,j in zip(x,y):
    print(i,j)
```

# Object Oriented Approach

```python
# - It is also known as figure and axes approach.
# - In this approach we crate a figure and then we create axes inside the figure.
# - The data is plotted inside the axes.
# - We can increase ordecrease the size of the figure.
```

### Creating a figure

```python
# plt.figure() - It returns a figure object.
# - The virtually created, we can not see it.
```

```python
f=plt.figure()
```

```python
print(type(f))
```

### Creating axes inside the figure

```python
# add_axes() - It is used to create an axes inside a figure.
# Note- The cell which we create a figure in, inside the same cell we have to create the axes
```

```python
f=plt.figure(figsize=(6,6))
a=f.add_axes([0.1,0.1,0.5,0.5])
b=f.add_axes([0.7,0.1,0.5,0.5])
c=f.add_axes([0.7,0.7,0.5,0.5])
d=f.add_axes([0.1,0.7,0.5,0.5])
plt.show()
```

```python
f=plt.figure()
a=f.add_axes([0.1,0.1,0.5,0.5])
b=f.add_axes([0.7,0.1,0.5,0.5])
c=f.add_axes([0.7,0.7,0.5,0.5])
d=f.add_axes([0.1,0.7,0.5,0.5])

a.plot(x,y)
d.plot(y,x)
c.plot(x**2,y/.25)
b.plot([1,2,3,4],[4,2,3,5])
plt.show()
```

### Giving the title, label inside the object oriented approach

```python
# axes_name.set_title('title')
# axes_name.set_xlabel('Label on x Axis')
# axes_name.set_ylabel('Label on y Axis')
```

```python
f=plt.figure()
a=f.add_axes([0.1,0.1,0.5,0.5])
b=f.add_axes([0.7,0.1,0.5,0.5])
c=f.add_axes([0.7,0.8,0.5,0.5])
d=f.add_axes([0.1,0.8,0.5,0.5])

a.plot(x,y)
a.set_title('Plot-A')
a.set_xlabel('Array-X')
a.set_ylabel('Array-Y')

d.plot(y,x)
d.set_title('Plot-D')
d.set_xlabel('Array-1')
d.set_ylabel('Array-2')

c.plot(x**2,y/.25)
b.plot([1,2,3,4],[4,2,3,5])
plt.show()
```

### Pie Chart

```python
# plt.pie(values,labels,autopct=,explode=)
# values=array/list
# labels=list/array
# autopct- format specifier
# explode=tuple/list
```

```python
money=[2000,1500,4000,3000]
names=['A','B','C','D']
```

```python
plt.pie(money,
        labels=names,
        autopct='%.2f%%',
       explode=[0.5,0.1,0.1,0.1],
       colors=['red','green','yellow','blue'],
       shadow=True,
       labeldistance=.9,
       pctdistance=1.25)

# plt.legend(loc=10)
plt.show()
```
