# How a Python program runs?

# Source Code

```python
# - The code that we write as a programmar is known as source code.
# - The file type of Python source code is .py
```

# Intermediate Code

```python
# - Once we execute the Python source code then compiler comes into picture.
# - The compiler reads the whole source code and converts it into intermediate code (if no error found)
# - The Python intermediate is also known as Bytecode.
# - The extension of the Bytecode file is .pyc.
# - If compiler catches any error in the source then it does not create any bytecode.
# - In this case no output is generated.
```

# Machine Code

```python
# - Once compiler has created the Bytecode then interpreter comes into picture.
# - Interpreter reads the Bytecode line by line and converts it into Machine code.
# - As soon as the interpreter catches any error it stops the execution.
# - In this case until the interpreter catches any error we get the output.
```
