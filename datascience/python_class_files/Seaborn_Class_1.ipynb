### displot()

```python
# sns.displot(x/y='numerical',hue='categorical',data=data_frame)
# kind='hist', 'kde'
# bins=int (applicale only when the kind is hist)
# stat=count, density, probability, frequency
```

```python
# plt.figure(figsize=(20,6),dpi=100)
sns.displot(x='age',data=tt,kind='hist',hue='pclass',bins=45,stat='frequency',
            height=8,aspect=2,color='green',palette='tab10')

plt.show()
```

```python
# plt.figure(figsize=(20,6),dpi=100)
sns.displot(y='age',data=tt,kind='hist',hue='pclass',bins=45,stat='frequency',
            height=8,aspect=2,color='green',palette='tab10')

plt.show()
```

### histplot()

```python
# sns.histplot(x/y=numerical, hue='categorical', data=data_frame, stat=, kde=True)
```

```python
plt.figure(figsize=(10,8),dpi=100)
sns.histplot(x='age',data=tt, kde=True, bins=35,stat='frequency',
             hue='pclass',palette='tab10')

plt.show()
```

### sns.barplot()

# Plotting the seaborn plots in  axes

```python
f,a=plt.subplots(2,2,figsize=(10,8),dpi=100)

sns.boxplot(y='age',data=tt,ax=a[0,0])
sns.scatterplot(x='age',y='fare',data=tt,ax=a[0,1])
sns.histplot(x='age',data=tt,ax=a[1,0])
sns.violinplot(x='sex',y='age',data=tt,ax=a[1,1])

plt.tight_layout()
plt.show()
```
