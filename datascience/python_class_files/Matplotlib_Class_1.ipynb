# Giving the title, axis names and the markers

```python
# plt.title('Title',fontdict={})
# plt.xlabel('Name to the x-axis')
# plt.ylabel('Name to the y-axis')
```

```python
plt.plot(x,y)
plt.title('Plot-1',fontdict={'color':'red','size':20,'font':'monospace'})
plt.xlabel('Array-X',fontdict={'color':'red','size':20,'font':'monospace'})
plt.ylabel('Array-Y',fontdict={'color':'red','size':20,'font':'monospace'})
plt.show()
```

# Markers

```python
# plot(marker='type of the marker')
# marker types= 'o', 'v', '*', 'p', '8', '<', '>'
```

```python
plt.plot(x,y,marker='o',
         markersize=12,
         markerfacecolor='yellow',
        markeredgecolor='red',
        markeredgewidth=2)
plt.show()
```

# Line styles and the linewidths

```python
# ls='-','--','-.',':'
# lw=int/float
```

```python
plt.plot(x,y,ls='-',lw=4)
plt.show()
```

# Color

```python
# 'red' -> 'r','green'-'g','blue'->'b','yellow'->'y','aqua','black'-> 'k'
```

```python
plt.plot(x,y,ls='-',lw=4, color='black')
plt.show()
```

# Legends

```python
# - Legends tell what is what
```

```python
plt.plot(x,y,
         lw=4, 
         color='red',
        label='Sales Trend')

# plt.legend(loc='center right')
plt.legend(loc=9,labelcolor='red',fontsize=15)
plt.show()
```

```python
plt.plot(x,y,
         color='green',
         marker='o',
         markersize=12,
         markerfacecolor='yellow',
        markeredgecolor='red',
        markeredgewidth=2,
        label='Sales Trend',
        lw=2.8,
        ls='--')

plt.title('Plot-1',fontdict={'color':'red','size':20,'font':'monospace'})
plt.xlabel('Array-X',fontdict={'color':'red','size':20,'font':'monospace'})
plt.ylabel('Array-Y',fontdict={'color':'red','size':20,'font':'monospace'})

plt.legend(fontsize=12,labelcolor='green')
plt.show()
```
