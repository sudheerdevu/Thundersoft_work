### how='inner'

```python
# - It takes the common observations for the common column from both the dataframes.
# - This is the default type
```

```python
# emp- 10,20,30,50
# dpt- 10,20,30,40
# inner- 10,20,30
```

```python
pd.merge(emp,dpt,on='Depart_No',how='inner')
```

```python
pd.merge(emp,dpt,on='Depart_No')
```

### how='left'

```python
# - It will take all the observations from the left dataframe from the common column
```

```python
# emp- 10,20,30,50
# dpt- 10,20,30,40
# left- 10,20,30,50
```

```python
pd.merge(emp,dpt,on='Depart_No',how='left')
```

### how='right'

```python
# - It will take all the observations from the right dataframe from the common column
```

```python
# emp- 10,20,30,50
# dpt- 10,20,30,40
# right- 10,20,30,40
```

```python
pd.merge(emp,dpt,on='Depart_No',how='right')
```

```python
emp
```

### how='outer'

```python
# It will take all the observations for the common column from both the dataframes 
```

```python
# emp- 10,20,30,50
# dpt- 10,20,30,40
# outer- 10,20,30,40,50
```

```python
pd.merge(emp,dpt,on='Depart_No',how='outer')
```

```python
pd.merge(emp,dpt,on=['Depart_No','City'],how='inner')
```

```python
def get_digit(x):
    m=re.search(r'[0-9.]+',x)
    if m:
        return m.group()
    else:
        return x
```

### Creating additional columns

```python
df=pd.DataFrame({'Name':['Jayanth','Ashish','Aditya','Nidhi','Megha','Suresh','Hema'],
             'Marks':[80,65,45,28,52,44,52]})
```

```python
df
```

```python
# 60 or >60- First
# 45>marks<60 - Second
```

```python
def remarks(marks):
    if marks>=60:
        return 'First'
    elif marks>=45:
        return 'Second'
    elif marks>=33:
        return 'Third'
    else:
        return 'Fail'
```

```python
remarks(35)
```

```python
df['Remarks']=df.Marks.apply(remarks)
```

```python
df
```
